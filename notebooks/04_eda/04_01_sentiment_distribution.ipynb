# 04_01 Sentiment Distribution Analysis

Explores the distribution of the Bitcoin Fear & Greed Index values
and sentiment regimes in the merged dataset.

**Dependencies:** functions from `src/sentiment_trader_analytics/`

```python
from pathlib import Path

import pandas as pd

from sentiment_trader_analytics.eda.summary_stats import (
    compute_descriptive_stats,
    compute_missingness_report,
    compute_outlier_summary,
)
from sentiment_trader_analytics.visualization.plots import (
    plot_sentiment_regime_frequency_barplot,
    plot_sentiment_value_histogram,
)
```

```python
df = pd.read_parquet("../../data/features/feature_store_20260701_103726.parquet")
print(f"Loaded {len(df)} rows, {len(df.columns)} columns")
```

## 1. Sentiment Value Distribution

```python
plot_sentiment_value_histogram(
    df, "../../outputs/figures/eda/sentiment_value_histogram.png"
)
```

## 2. Sentiment Regime Frequency

```python
plot_sentiment_regime_frequency_barplot(
    df, "../../outputs/figures/eda/sentiment_regime_frequency_barplot.png"
)
```

## 3. Descriptive Statistics

```python
stats_columns = ["sentiment_value", "Closed PnL", "Leverage"]
desc = compute_descriptive_stats(df, stats_columns)
display(desc)
```

## 4. Missingness Report

```python
missing = compute_missingness_report(df)
display(missing)
```

## 5. Outlier Summary (Sentiment Value)

```python
outlier = compute_outlier_summary(df, "sentiment_value")
print(f"Outliers: {outlier.n_outliers} / {len(df)} ({outlier.n_outliers/len(df)*100:.1f}%)")
print(f"Bounds: [{outlier.lower_bound:.2f}, {outlier.upper_bound:.2f}]")
```

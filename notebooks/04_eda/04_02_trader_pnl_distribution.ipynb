# 04_02 Trader PnL Distribution Analysis

Explores the distribution of trader Closed PnL and PnL by
sentiment regime.

**Dependencies:** functions from `src/sentiment_trader_analytics/`

```python
from pathlib import Path

import pandas as pd

from sentiment_trader_analytics.eda.summary_stats import (
    compute_descriptive_stats,
    compute_outlier_summary,
)
from sentiment_trader_analytics.visualization.plots import (
    plot_pnl_by_sentiment_boxplot,
    plot_trader_pnl_distribution_histogram,
)
```

```python
df = pd.read_parquet("../../data/features/feature_store_20260701_103726.parquet")
print(f"Loaded {len(df)} rows, {len(df.columns)} columns")
```

## 1. Trader PnL Distribution

```python
plot_trader_pnl_distribution_histogram(
    df, "../../outputs/figures/eda/trader_pnl_distribution_histogram.png"
)
df["Closed PnL"].describe()
```

## 2. PnL by Sentiment Regime (Box Plot)

```python
plot_pnl_by_sentiment_boxplot(
    df, "../../outputs/figures/eda/pnl_by_sentiment_boxplot.png"
)
```

## 3. PnL by Regime Descriptive Stats

```python
pnl_stats = (
    df.dropna(subset=["sentiment_classification", "Closed PnL"])
    .groupby("sentiment_classification", observed=True)["Closed PnL"]
    .agg(["count", "mean", "std", "min", "median", "max"])
)
display(pnl_stats)
```

## 4. Outlier Summary (Closed PnL)

```python
outlier = compute_outlier_summary(df, "Closed PnL", method="iqr", config={"iqr_multiplier": 1.5})
print(f"Outliers: {outlier.n_outliers} / {len(df)} ({outlier.n_outliers/len(df)*100:.1f}%)")
print(f"Bounds: [{outlier.lower_bound:.2f}, {outlier.upper_bound:.2f}]")
```

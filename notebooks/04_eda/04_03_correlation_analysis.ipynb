# 04_03 Correlation Analysis

Explores correlations between sentiment values and trader behavior
metrics (PnL, leverage, win rate, trade count).

**Dependencies:** functions from `src/sentiment_trader_analytics/`

```python
import numpy as np
import pandas as pd

from sentiment_trader_analytics.eda.summary_stats import (
    compute_descriptive_stats,
    compute_missingness_report,
)
from sentiment_trader_analytics.visualization.plots import (
    plot_feature_correlation_heatmap,
    plot_missingness_heatmap,
)
```

```python
df = pd.read_parquet("../../data/features/feature_store_20260701_103726.parquet")
print(f"Loaded {len(df)} rows, {len(df.columns)} columns")
```

## 1. Missingness Heatmap

```python
plot_missingness_heatmap(
    df, "../../outputs/figures/eda/missingness_heatmap.png"
)
```

## 2. Missingness Report

```python
missing = compute_missingness_report(df)
display(missing.head(10))
```

## 3. Feature Correlation Heatmap

```python
numeric_features = [
    "sentiment_value",
    "Closed PnL",
    "Leverage",
    "trader_win_rate_7d",
    "trader_pnl_rolling_7d",
    "trader_pnl_rolling_30d",
    "trader_leverage_avg_24h",
    "trader_pnl_volatility_14d",
    "trader_trade_count_7d",
    "trader_avg_size_usd_7d",
]
plot_feature_correlation_heatmap(
    df,
    "../../outputs/figures/eda/feature_correlation_heatmap.png",
    numeric_features=numeric_features,
    method="spearman",
)
```

## 4. Descriptive Statistics

```python
desc = compute_descriptive_stats(df, numeric_features)
display(desc)
```
